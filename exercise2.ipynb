# GCD

```python
import pyscipopt as scip
import numpy as np
```

```python
model = scip.Model('Greatest Common Divisior')
```

```python
a = int(np.random.rand() * 100)
b = int(np.random.rand() * 100)
```

```python
x = model.addVar(vtype = 'I', lb=-1000 , name = 'x')
y = model.addVar(vtype = 'I', lb=-1000 , name = 'y')
```

```python
constraint1 = model.addCons(a * x + b * y >= 1, name = 'x_pos')
```

```python
model.setObjective(a * x + b * y)
```

```python
model.optimize()
```

```python
print(f"The model status is '{model.getStatus()}', solved after {model.getSolvingTime()} seconds.")
print(f"The gcd of {a} and {b} is {model.getObjVal()}.")
print(f"It can be represented as gcd({a}, {b}) = {model.getObjVal()} = {a} * {model.getVal(x)} + {b} * {model.getVal(y)}.")
```

```python
def EuclidAlgorithm(a,b,t = 1):
    if t == 1:
        a,b = max(a,b), min(a,b)

    if a == 0:
        return b
    if b == 0:
        return a
    
    return EuclidAlgorithm(b, a % b, 0)
```

```python
def gcdByIntegerSol(a_list,b_list):

    sol = []
    for index, (a,b) in enumerate(zip(a_list, b_list)):

        model = scip.Model(f'GCD{index}')
        
        x = model.addVar(vtype = 'I', lb = -1000, name = 'x')
        y = model.addVar(vtype = 'I', lb = -1000, name = 'y')

        model.addCons(a * x + b * y >= 1, name = 'positivCons')
        model.setObjective(a * x + b * y)

        model.hideOutput()  
        model.optimize()

        sol.append(model.getObjVal())
    
    return sol
```

```python
my_results = gcdByIntegerSol(
    [13, 8265, 918, 2021, 36, 150, -20],
    [2 , 24,   2781,-31,  81, 35,  -96]
)
```
